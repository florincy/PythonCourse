### Introdução ao matplotlib

O **matplotlib** é uma biblioteca  do Python para criar gráficos 2D de maneira simples. Abaixo, são descritos alguns comandos básicos para criar gráficos com essa biblioteca, a variedade de comandos está ligada à flexibilidade ofertada por essa biblioteca. Usaremos tais comandos para mostrar nossas regressões lineares.

### Figura e Tamanho

- **`plt.figure(figsize=(largura, altura))`**:  
  Este comando define o tamanho da figura em polegadas, ajustando a escala do gráfico de acordo com as dimensões especificadas.  

### Adição de Pontos e Linhas

- **`plt.scatter(x, y, color='cor', label='legenda')`**:  
  Plota pontos baseados nas coordenadas `x` e `y`, permitindo customizar cor e legenda.

- **`plt.plot(x, y, color='cor', label='legenda')`**:  
  Conecta os valores de `x` e `y` com uma linha, sendo frequentemente usado para adicionar ajustes, como curvas de regressão.

### Adicionando Informações ao Gráfico

- **`plt.title("Título")`**:  
  Define o título do gráfico.

- **`plt.xlabel("Nome do eixo X")`** e **`plt.ylabel("Nome do eixo Y")`**:  
  Adicionam rótulos aos eixos X e Y.

- **`plt.legend()`**:  
  Exibe uma legenda explicando os elementos presentes no gráfico.

### Estética e Interatividade

- **`plt.grid(True)`**:  
  Adiciona uma grade ao fundo do gráfico, tornando a leitura mais intuitiva.

### Exibição

- **`plt.show()`**:  
  Renderiza o gráfico na tela, sendo essencial para exibir o resultado quando várias figuras são criadas em sequência.






```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

#DADOS DO ARTIGO: https://doi.org/10.3390/molecules25010024

#Dados
data = {
    "PIC50": [12.94,13.23,13.12,12.45,11.21,10.87,9.47,14.43,9.33,11.27,12.94],
    "Dipole_Moment": [5.636, 5.646, 5.610, 5.575, 5.475, 5.099, 9.575, 6.154, 6.567, 4.510, 6.952],
    "PSA": [87.38, 91.98, 96.58, 101.18, 105.55, 102.75, 73.94, 73.94, 88.38, 73.94, 69.18],
    "Softness": [3.615, 4.613, 3.614, 3.576, 3.535, 3.502, 3.427, 4.528, 3.612, 3.685, 4.011]

}

#Converte os dados do formato de lista para o formato de dataframe
df = pd.DataFrame(data)

# Regressão para Momento de dipolo vs log(IC50)
dipole_model = LinearRegression()
dipole_model.fit(df[['Dipole_Moment']], df['PIC50'])
dipole_pred = dipole_model.predict(df[['Dipole_Moment']])

# Regressão para TPSA vs log(IC50)
tpsa_model = LinearRegression()
tpsa_model.fit(df[['PSA']], df['PIC50'])
tpsa_pred = tpsa_model.predict(df[['PSA']])

# Regressão para TPSA vs Softness
softness_model = LinearRegression()
softness_model.fit(df[['Softness']], df['PIC50'])
softness_pred = softness_model.predict(df[['Softness']])

# Constrói uma regressão linear múltipla usando o Momento de Dipolo, PSA e o caráter mácio
X = df[['Dipole_Moment', 'PSA','Softness']]  # Variáveis independentes (X) -> Vetor de variáveis independentes
y = df['PIC50']  # Variável predita (Y)
model = LinearRegression()
model.fit(X, y)

# Gráfico 1: Momento de dipolo vs log(IC50)
plt.figure(figsize=(8, 6))
plt.scatter(df['Dipole_Moment'], df['PIC50'], color='pink', alpha=0.7, label='Conjunto de dados')
plt.plot(df['Dipole_Moment'], dipole_pred, color='red', label='Regressão linear')
plt.title("Momento de dipolo vs p(IC50)")
plt.xlabel("Momento de dipolo")
plt.ylabel("p(IC50)")
plt.legend()
plt.grid(True)  # Adiciona uma grade ao gráfico
plt.show()

# Gráfico 2: PSA vs log(IC50)
plt.figure(figsize=(8, 6))
plt.scatter(df['PSA'], df['PIC50'], color='green', alpha=0.7, label='Conjunto de dados')
plt.plot(df['PSA'], tpsa_pred, color='red', label='Regressão linear')
plt.title("PSA vs p(IC50)")
plt.xlabel("PSA")
plt.ylabel("p(IC50)")
plt.legend()
plt.grid(True)
plt.show()

# Gráfico 3: Softness vs IC50
plt.figure(figsize=(8, 6))
plt.scatter(df['Softness'], df['PIC50'], color='purple', alpha=0.7, label='Conjunto de dados')
plt.plot(df['Softness'], softness_pred, color='red', label='Regressão linear')
plt.title("Softness vs p(IC50)")
plt.xlabel("Softness")
plt.ylabel("p(IC50)")
plt.legend()
plt.grid(True)
plt.show()


# Exibe os coeficientes do modelo de regressão linear múltipla
print("\nCoeficientes da Regressão Linear Múltipla\n")
print(f"Termo independente: {model.intercept_}")
print(f"Demais coeficientes: {model.coef_}")

# Calcula o R^2 para regressão multilinear
y_pred = model.predict(X)
r2 = r2_score(y, y_pred)
print(f"\nR-quadrado: {r2}")
```
